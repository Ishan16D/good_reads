### This script creates my base books list

```python
import requests as re
from bs4 import BeautifulSoup
```

```python
# Getting the urls in the strong female character list

url = 'https://www.goodreads.com/list/show/1023.Best_Strong_Female_Fantasy_Novels'
pc = 35
sfc_urls = []
urls_shared = []
books_titles = []
base_url = 'https://www.goodreads.com'

for i in range(pc + 1):
    if i != 0:
        r = re.get(url + f'?page={i}')
        soup = BeautifulSoup(r.content, 'html.parser')
        for book in soup.find_all('a', attrs = {'class' : 'bookTitle'}):
            sfc_urls.append(base_url + book.attrs['href'])
        print(f'Finished with page {i} out of {pc}')
```

```python
# Getting urls for most anticipated 2013-2021

list_urls = ['https://www.goodreads.com/list/show/20917', 'https://www.goodreads.com/list/show/38755',
        'https://www.goodreads.com/list/show/74206.Can_t_Wait_Sci_Fi_Fantasy_of_2015',
        'https://www.goodreads.com/list/show/89486', 'https://www.goodreads.com/list/show/100366',
        'https://www.goodreads.com/list/show/112840', 'https://www.goodreads.com/list/show/124876',
        'https://www.goodreads.com/list/show/134406', 'https://www.goodreads.com/list/show/142995']
pcs = [7, 5, 6, 5, 4, 6, 6, 5, 2]
books_urls = []

for url in list_urls:
    for page in range(pcs[list_urls.index(url)] + 1):
        if page != 0:
            r = re.get(url + f'?page={page}')
            soup = BeautifulSoup(r.content, 'html.parser')
            for book in soup.find_all('a', attrs = {'class' : 'bookTitle'}):
                books_urls.append(base_url + book.attrs['href'])
            print(f'Finished with page {page} out of {len(pcs)} for {url}')

            
            
```

```python
urls_shared = [url for url in books_urls if url in sfc_urls]
```

```python
books_titles =[]
series_list = []
romance_novels = []
epic_fantasy_novels = []

for url in urls_shared:
    try:
        r = re.get(url)
        soup = BeautifulSoup(r.content, 'html.parser')
        genres = []
        for genre in soup.find_all('a', attrs={'class' : 'bookPageGenreLink'}):
            if '/genres/' in str(genre.attrs['href']):
                genres.append(str(genre.attrs['href'])[8:])
        if 'romance' not in genres:
            if str(soup.find_all('a', attrs={'class' :'greyText'})[0].string) != 'Edit Details':
                series = str(soup.find_all('a', attrs={'class' :'greyText'})[0].string[10:-5])
            else:
                print(f'{url} is not a series')
            if series not in series_list:
                series_list.append(series)
                title = str(soup.find_all('h1', attrs={'id' : 'bookTitle'})[0].string)[7:-1]
                books_titles.append(title)
            else:
                print(f'{series} already done')
        elif 'romance' not in genres and 'epic-fantasy' in genres:
            if str(soup.find_all('a', attrs={'class' :'greyText'})[0].string) != 'Edit Details':
                series = str(soup.find_all('a', attrs={'class' :'greyText'})[0].string[10:-5])
            else:
                print(f'{url} is not a series')
            if series not in series_list:
                series_list.append(series)
                title = str(soup.find_all('h1', attrs={'id' : 'bookTitle'})[0].string)[7:-1]
                epic_fantasy_novels.append(title)
                print(f'{title} is an epic fantasy novel')
            else:
                print(f'{series} already done')
        else:
            if str(soup.find_all('a', attrs={'class' :'greyText'})[0].string) != 'Edit Details':
                series = str(soup.find_all('a', attrs={'class' :'greyText'})[0].string[10:-5])
            else:
                print(f'{url} is not a series')
            if series not in series_list:
                series_list.append(series)
                title = str(soup.find_all('h1', attrs={'id' : 'bookTitle'})[0].string)[7:-1]
                romance_novels.append(title)
                print(f'{title} is a romance novel')
            else:
                print(f'{series} already done')
    except IndexError:
        print(f'{url} failed')
        
        
    print(f'Finished book {urls_shared.index(url)} out of {len(urls_shared)}')
                  
print(len(books_titles))
print(len(epic_fantasy_novels))
print(len(romance_novels))
        
    
         
```

```python
all_books = books_titles

print(len(all_books))

for book in romance_novels:
    all_books.append(book + ' (R)')

print(len(all_books))
```

```python
f = open('base_books.txt', 'a')

for book in all_books:
    f.write(book)
    f.write('\n')
    
f.close()
```
